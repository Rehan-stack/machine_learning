```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
from sklearn import datasets
```

```python

df = pd.read_csv("carprices.csv")
df
```

```python
plt.scatter(df[['Mileage']],df[['Sell Price($)']],color='red')
plt.xlabel('mileage')
plt.ylabel('price')
```

```python
plt.scatter(df[['Age(yrs)']],df[['Sell Price($)']],color='red')
plt.xlabel('age per year')
plt.ylabel('price')
```

```python
dummies_variables = pd.get_dummies(df[['Car Model']])
dummies_variables
```

```python
merge = pd.concat([df,dummies_variables],axis='columns')
merge
```

```python
final = merge.drop(['Car Model','Car Model_BMW X5'],axis='columns')
final
```

```python
X = final.drop(['Sell Price($)'],axis='columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
model = linear_model.LinearRegression()
model
```

```python
model.fit(X,y)
```

```python
model.predict([[50000,6,0,0]])
```

```python
model.score(X,y)
```
